```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_point(side:float = 2.0, num_points:int=10000):
  global out_x, out_y, cir_x, cir_y,sq_x, sq_y
  sq_x = (side*np.random.rand(num_points))- side/2
  sq_y = (side*np.random.rand(num_points))- side/2
  mask = (sq_x**2 + sq_y**2 <= (side/2)**2)
  cir_x = sq_x[mask]
  cir_y = sq_y[mask]
  out_x = sq_x[~mask]
  out_y = sq_y[~mask]
  return len(cir_x)/len(sq_x)*side**2
```

```python
def draw_plot(side:float=2.0):
  fig,ax = plt.subplots(1)
  ax.scatter(out_x,out_y, color = "red")
  ax.scatter(cir_x,cir_y,color = "skyblue")
  ax.add_artist(plt.Circle((0, 0), side/2, alpha=0.5, color='white'))
  ax.set_aspect('equal')
  plt.xlim((-side/2, side/2)); plt.ylim((-side/2, side/2))
  plt.show()
```

```python
side = 2.0
x = generate_point(side, num_points=10000)
print(x)
draw_plot(side)
```

```python
X = np.arange(1,10000)
y = [np.pi- generate_point(num_points=x) for x in X]
plt.plot(X,y)
plt.ylabel("Error")
```
